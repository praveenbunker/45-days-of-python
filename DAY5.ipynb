```python
def square(x):
  return x*x
#map(minimum 2 argument)

l = [10,20,30]
l1 = list(map(square,l))
print(l1)


#alternate without in built fxn
l1= []
for i in range(len(l)):
  l1.append(square(l[i]))
print(l1)


#filter(minimum 2 argument)
def hello(x):

  return x.startswith("a")


l = ["apple","banana","cat","dog"]
l1 = list(filter(hello,l))
print(l1)


#example

l1 = []
for i in range(len(l)):
  if(hello(l[i])):
    l1.append(l[i])
print(l1)


#examle change in upper case

l = ["apple","banana","cat","dog"]
l1 = list(map(lambda x:x.upper(),l))
print(l1)


#reduce(minimum 2 argument)
l = [10,20,30]

```
