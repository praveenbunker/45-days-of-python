## Multiple line chart and legends

```python
import matplotlib.pyplot as plt

x = [2010,2015,2020,2025]
y1 = [100,200,260,290]
y2 = [150,180,215,330]

plt.plot(x,y1,label="jeans")
plt.plot(x,y2,label="T-shirt")
plt.legend()
plt.show()
```

```python
#questio multiple line chart


import matplotlib.pyplot as plt

x = [2000, 2005, 2010, 2015]
y1 = [110, 140, 150, 300]
y2 = [100, 321, 150, 653]
y3 = [200,400,100,50]

plt.plot(x, y1,
         label="jeans",
         color="green",
         marker="o",
         linestyle="--",
         linewidth=4,
         markersize=10)

plt.plot(x, y2,
         label="T-shirt",
         color="pink",
         marker="*",
         linestyle="-",
         linewidth=4,
         markersize=10)

plt.plot(x, y3,
         label="shirt",
         color="red",
         marker="o",
         linestyle=":",
         linewidth=2,
         markersize=10)


plt.legend()
plt.show()
```

## ***Bar chart***

```python
import matplotlib.pyplot as plt
x = [2015,2020,2025,2030]
y = [100,150,200,290]

plt.figure(figsize=(6,2))
plt.bar(x,y,color="red")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

groups = ["group A", "group B", "group C", "group D"]
y1 = [10, 20, 20, 40]
y2 = [20, 30, 25, 30]

x = np.arange(len(groups))
width = 0.4

plt.bar(x - width/2, y1, width, label="boys", color="red")
plt.bar(x + width/2, y2, width, label="girls", color="blue")

plt.xlabel("groups")
plt.ylabel("number of students")
plt.title("No. of students in each group")
plt.xticks(x, groups)
plt.legend()

plt.show()
```
