```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# E-Commerce Dataset
data = {
    "product": ["Laptop", "Mobile", "Headphone", "Tablet", "Smartwatch", "Camera"],
    "price": [50000, 20000, 3000, np.nan, 5000, 25000],
    "quantity": [10, 25, 40, 15, 30, 12]
}

df = pd.DataFrame(data)

# NaN values fill (FutureWarning fixed)
df["price"] = df["price"].fillna(df["price"].mean())

# Revenue column
df["revenue"] = df["price"] * df["quantity"]

print("E-Commerce Data:")
print(df)

# NumPy Operations
print("\nAverage Revenue :", np.mean(df["revenue"]))
print("Maximum Revenue :", np.max(df["revenue"]))
print("Minimum Revenue :", np.min(df["revenue"]))

# 2D Subplots (2 rows, 2 columns)
fig, aux = plt.subplots(2, 2, figsize=(12, 8))

# 1. Line Chart
aux[0, 0].plot(df["product"], df["revenue"], marker="o")
aux[0, 0].set_title("Product vs Revenue")
aux[0, 0].set_xlabel("Product")
aux[0, 0].set_ylabel("Revenue")

# 2. Bar Chart
aux[0, 1].bar(df["product"], df["quantity"])
aux[0, 1].set_title("Product vs Quantity Sold")
aux[0, 1].set_xlabel("Product")
aux[0, 1].set_ylabel("Quantity")

# 3. Scatter Plot
aux[1, 0].scatter(df["price"], df["quantity"])
aux[1, 0].set_title("Price vs Quantity")
aux[1, 0].set_xlabel("Price")
aux[1, 0].set_ylabel("Quantity")

# 4. Pie Chart
aux[1, 1].pie(
    df["revenue"],
    labels=df["product"],
    autopct="%1.1f%%"
)
aux[1, 1].set_title("Revenue Distribution")

plt.tight_layout()

# Save PDF
plt.savefig("ecommerce_project.pdf")

plt.show()
```
